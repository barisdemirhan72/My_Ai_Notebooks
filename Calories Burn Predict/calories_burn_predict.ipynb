# Calories Burn Predict ML Model

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("calorie_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna
```

```python
df.columns = ["ID","Gender","Age","Height(m)","Weight(kg)","BMR","Activity Level","TDEE"]
```

```python
df.head()
```

```python
df.drop(columns=["ID"])
```

```python
pd.get_dummies(df,columns=["Activity Level"])
```

```python
pd.get_dummies(df,columns=["Gender"])
```

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df.head()
```

```python
df['Activity Level'] = df['Activity Level'].map({'Sedentary': 1, 'Lightly Active': 2, 'Moderately Active': 3, 'Very Active': 4, 'Super Active': 5})
```

```python
df.head()
```

```python
df.dropna
```

```python
df.drop(columns=["ID"])
```

```python
x = df[["Gender","Age","Height(m)","Weight(kg)","Activity Level"]]
```

```python
y = df[["TDEE"]]
```

```python
scaler_model = StandardScaler()
x_scaled = scaler_model.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,train_size=0.80,random_state=48)
```

```python
model = RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_predict)
print(f"Mean Absolute Error (MAE): {mae:.2f} kalori")
```
